# 01 Vector Similarity

目标：理解 dot product 和 cosine similarity 为什么可以用于 RAG 中的向量检索。

完成标准：手算一个小例子，用 NumPy 和 PyTorch 复现，并解释它和 query/chunk 检索的关系。

## 手算例子

设：

- a = [1, 2, 3]
- b = [2, 1, 0]

先手算 dot product，再手算 cosine similarity。

```python
import numpy as np

a_np = np.array([1, 2, 3], dtype=float)
b_np = np.array([2, 1, 0], dtype=float)

dot_np = np.dot(a_np, b_np)
cosine_np = dot_np / (np.linalg.norm(a_np) * np.linalg.norm(b_np))

dot_np, cosine_np
```

```python
import torch

a_torch = torch.tensor([1.0, 2.0, 3.0])
b_torch = torch.tensor([2.0, 1.0, 0.0])

dot_torch = torch.dot(a_torch, b_torch)
cosine_torch = dot_torch / (torch.norm(a_torch) * torch.norm(b_torch))

dot_torch, cosine_torch
```

```python
print("a shape:", a_torch.shape)
print("b shape:", b_torch.shape)
print("dot shape:", dot_torch.shape)
```
