## Linear Algebra part

### Linear Regression

$y = wx + b$


```python
import numpy as np
import matplotlib.pyplot as plt

# ==== 1.造一些数据=====

np.random.seed(0)
x = np.linspace(0,10,20)
y = 3 * x + 5 + np.random.randn(20)

# ==== 2.拟合一条直线=====
w, b = np.polyfit(x,y,1) # 1表示一次函数

y_pred = w * x + b

# ==== 3.画图=====
plt.figure(figsize=(8,5))

#原始点
plt.scatter(x,y,label="data points")

#回归线
plt.plot(x,y_pred,color ="red",label=f"fit line:y={w:.2f}x+{b:.2f}")

plt.title("Linear Regression Visualization")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
